## 1. PRESENTATION DAY MANAGEMENT

### 1.1. Time Management
*   **Duration:** 10 Minutes Presentation + 5 Minutes Q&A.

### 1.2. Presentation Format
**Friends, I do not want you to project code onto the screen and read it line by line.** Tell me the story.

*   **Wrong:** "I did a `dropna` here, then I did a `groupby`..." (Boring).
*   **Right:** "There was 20% missing data here. Since the data was normally distributed, I filled it with the mean. As a result, I obtained this chart..." (Impressive/Engaging).
